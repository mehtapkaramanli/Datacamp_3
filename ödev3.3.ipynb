```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
from scipy import stats
import seaborn as sns

plt.figure(figsize=(8,4), dpi = 100)
orneklem_ornek = np.random.normal(100, 400, 5000)
orneklem_normal = np.random.normal(np.mean(orneklem_ornek),
                                         np.std(orneklem_ornek), 5000)
plt.hist(orneklem_normal, bins = 50)
plt.show()
```

```python
orneklem_normal.sort()
orneklem_ornek.sort()
```

```python
plt.figure(figsize=(8,5), dpi = 200)
baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }
eksen_font  = {'family': 'arial', 'color': 'darkblue','weight': 'bold','size': 10 }
```

```python
plt.title('Örnek Veri Kümesi & Normal Dağılımlı Veri Kümesi', fontdict = baslik_font)
plt.xlabel('Örnek Veri Kümesi', fontdict = eksen_font)
plt.ylabel('Normal Dağılımlı Veri Kümesi', fontdict = eksen_font)
plt.scatter(orneklem_normal, orneklem_ornek)
plt.plot([np.min(orneklem_normal), np.max(orneklem_normal)],
         [np.min(orneklem_ornek), np.max(orneklem_ornek)], c='red')
```

```python
res = stats.probplot(orneklem_ornek, plot=plt)
plt.title("Q-Q Çizgesi", fontdict = baslik_font)
plt.xlabel("Teorik Değerler", fontdict = eksen_font)
plt.ylabel("Örnek Veriler", fontdict = eksen_font)
plt.show()
```

 Veri kümesi normal dağılıma yakın görünüyor. Çünkü düz çizgiden sapmalar az.
